```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```
df_casos = pd.read_excel(casosb_xlsx)
```

```
df_casos
```

```
Xc = df_casos.iloc[:, 0].values
yc = df_casos.iloc[:, 1].values
```

```
#Correla√ß√£o
np.corrcoef(Xc, yc)
```

```
Xc = Xc.reshape(-1, 1)
```

```
# REGRESS√ÉO LINEAR
regr_c = LinearRegression()
regr_c.fit(Xc, yc)

# Y = a + bx
#Inclina√ß√£o Variavel B
incl_c = regr_c.coef_
#Intercepta√ß√£o - Constante A
intercep_c = regr_c.intercept_
```

```
# PLOTTER
plt.title('RL COVID19 - Brazil - 20200409')
plt.xlabel('Dias')
plt.ylabel('Casos novos')
plt.scatter(Xc, yc) 
plt.plot(Xc, regr_c.predict(Xc), color='red')
plt.savefig('RL COVID19b - Brazil - 20200409.png', format='png')
```

```
# FORMULA LINEAR
variavel_casos = 30
previsao_casos = intercep_c + (incl_c * variavel_casos)
print(previsao_casos) 
```

```
#Score Regressor
regr_c.score(Xc, yc)
```

```
# ACUMULADOR EM X DIAS
descobrir = 30
df_casos.columns   #Colunas
soma_casos = df_casos['casos novos'].sum()
#print(soma_casos)
conta_df_casos = df_casos['dias corridos'].count()

acumular = soma_casos
#FOR para acumular
for c in range(conta_df_casos + 1, descobrir):
  prever = intercep_c + (incl_c * c)
  acumular += prever

print(int(acumular))
```

```
# SALVAR ACUMULA√á√ÉO EM NOVA PLANILHA - DIA A DIA

#for para add linhas coluna 'dias corridos'
lista_dias = list()
for dias in range(1, conta_df_casos + 1):
  lista_dias.append(dias)


#for para add linhas coluna 'casos novos RL'
lista_casos = list()
for casos in range(0, conta_df_casos):
  lista_casos.append(df_casos['casos novos'].values[casos])


#for para previs√£o
dias_aprever = 30 # COLOCAR O NUMERO DE DIAS TOTAL PARA PREVIS√ÉO ACUMULADA
contadorsimples = 0
for RL_dias in range(conta_df_casos +1, dias_aprever +1):
  lista_dias.append(RL_dias)
  calculo_previsao = intercep_c + (incl_c * RL_dias)
  '''int para transformar float em inteiros e [0] para pegar somente valores
  ao inv√©s de pegar array([numero])'''
  lista_casos.append(int(calculo_previsao[0]))
  contadorsimples += 1
#print(lista_casos)

#Novo df
new_df = pd.DataFrame({'dias corridos': lista_dias, 'casos novos RL': lista_casos})
new_df
```

```
# SALVAR EM EXCEL
new_df.to_excel('20200409-CASOS-RL30dias.xlsx', index=False, sheet_name='RL-Lucas', float_format='%.0f')
```
